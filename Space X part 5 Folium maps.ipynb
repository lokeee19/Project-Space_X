```python
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
url = "spacex_launch_geo.csv"
spacex_geo = pd.read_csv(url)
spacex_geo.head(35)
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_geo[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=500, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# For each launch site in the DataFrame, add a Circle object based on its coordinates and a Popup label
for _, site in launch_sites_df.iterrows():
    name = site["Launch Site"]
    latitude = site["Lat"]
    longitude = site["Long"]
    
    # Create a Circle object with the launch site's coordinates
    folium.Circle(
        location=[latitude, longitude],
        radius=1000,  # Adjust the radius as needed (in meters)
        color='blue',  # Circle color
        fill=True,
        fill_color='blue',  # Fill color for the circle
        fill_opacity=0.6,
        popup=name,  # Popup label with the launch site name
    ).add_to(site_map)

# Display the map
site_map
```

```python
import folium
from folium.plugins import MousePosition, MarkerCluster
from math import sin, cos, sqrt, atan2, radians

# Assuming you have the 'spacex_df' DataFrame with launch records
# Replace this DataFrame with your actual data
# spacex_df = ...

# Function to calculate distance between two points (launch site and clicked point)
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Create the map centered at the initial coordinate
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a marker cluster
marker_cluster = MarkerCluster()

# For each launch record in the DataFrame, create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successful or failed
for index, record in spacex_df.iterrows():
    name = record["Launch Site"]
    latitude = record["Lat"]
    longitude = record["Long"]
    launch_class = record["class"]
    
    # Set marker color based on the launch class
    marker_color = 'green' if launch_class == 1 else 'red'
    
    # Create a Marker with the launch site's coordinates and set the appropriate color
    marker = folium.Marker(
        location=[latitude, longitude],
        popup=name,  # Popup label with the launch site name
        icon=folium.Icon(color='white', icon_color=marker_color)  # Customize the marker color
    )
    
    # Add the Marker to the marker cluster
    marker_cluster.add_child(marker)

# Add the marker cluster to the site map
site_map.add_child(marker_cluster)

# Add MousePosition control to get the coordinates of the mouse over the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)


```

```python
# find coordinate of the closet coastline
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Given coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.57756
coastline_lat = 28.56324
coastline_lon = -80.5678

# Calculate the distance between the launch site and the coastline
distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance between launch site and coastline: {distance_to_coastline:.2f} km")
```

```python
# Create the map centered at the initial coordinate
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add a folium.Marker for the selected closest coastline point
folium.Marker(
    location=[coastline_lat, coastline_lon],
    popup="Closest Coastline Point",  # Popup label for the marker
).add_to(site_map)

# Create the icon for displaying the distance on the map
distance_icon = folium.DivIcon(
    icon_size=(20, 20),
    icon_anchor=(0, 0),
    html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_to_coastline:.2f} KM</b></div>'
)

# Add a folium.Marker with the distance icon to the launch site location
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup="Launch Site",  # Popup label for the marker
    icon=distance_icon,  # Use the custom icon for distance display
).add_to(site_map)

# Display the map
site_map
```
